```python
import requests
from bs4 import BeautifulSoup
```

```python
r=requests.get('https://www.pedigree.com.au/dog-breed-information/dog-breed-gallery/')
```

```python
r.text[:200]
```

```python
s=BeautifulSoup(r.text,'lxml')
type(s)
```

```python
d=s.find('div', {'id':'dogAZ'})
```

```python
d
```

```python
a_tags=d.find_all('a')
```

```python
type(a_tags)
```

```python
a_tags[:3]
```

```python
a_tags[3]
```

```python
a_tags[3]['href']
```

```python
a_tags[3].find('img')
```

```python
a_tags[3].find('img')['alt']
```

```python
a_tags[3].find('img')['src']
```

```python
rows=[]
url_prefix='https://www.pedigree.com.au'
for e in a_tags:
    img=e.find('img')
    breed=img['alt']
    src=url_prefix+img['src']
    img_html=f'<img src={src} />'
    link=url_prefix+e['href']
    href_html=f'<a href="{link}">{breed}</a>'
    rows.append((breed, link, src, href_html, img_html))
```

```python
rows
```

```python
df=pd.DataFrame(rows, columns=['breed', 'link', 'img_url', 'href', 'img_html'])
df[:3]
```
